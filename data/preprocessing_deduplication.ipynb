```python
import csv
import googletrans
from tqdm import tqdm
import time
import torch
```

```python
rows1 = []
# Load dataset1
with open('./dataset1/label.csv', 'r') as f:
    csv_data = csv.reader(f)
    header = next(csv_data)
    rows1 = list(csv_data)

rows2 = []
# Load dataset2
with open('./dataset2/label.csv', 'r') as f:
    csv_data = csv.reader(f)
    header = next(csv_data)
    rows2 = list(csv_data)
```

```python
# 공백 제거 & 오타 수정
def preprocess(tag_ko):
    tag_ko = tag_ko.replace(" ","")
    tag_ko = tag_ko.replace("ㅐㄴ드마크","랜드마크")
    tag_ko = tag_ko.replace("ㅏㅅ랑","사랑")
    tag_ko = tag_ko.replace("ㅑ경","야경")
    tag_ko = tag_ko.replace("ㅁ나세","만세")
    return tag_ko
```

```python
rows1
```

```python
# class 확인
tags_ko_all = []
rows = [rows1]
for idx, r in enumerate(rows):
    for i in range(0, len(r)):
        row = r[i]
        for tag_ko in row[2:]:
            if tag_ko != '':
                tag_ko = preprocess(tag_ko)
                tags_ko_all.append(tag_ko)

print("dataset1")
print("중복 제거되지 않는 tags list :", len(tags_ko_all))
tags_ko_all = sorted(list(set(tags_ko_all)))
print("중복 제거된 tags list ko :", len(tags_ko_all))
```

```python
# class 확인
tags_ko_all = []
rows = [rows2]
for idx, r in enumerate(rows):
    for i in range(0, len(r)):
        row = r[i]
        for tag_ko in row[2:]:
            if tag_ko != '':
                tag_ko = preprocess(tag_ko)
                tags_ko_all.append(tag_ko)

print("dataset2")
print("중복 제거되지 않는 tags list :", len(tags_ko_all))
tags_ko_all = sorted(list(set(tags_ko_all)))
print("중복 제거된 tags list ko :", len(tags_ko_all))
```

```python
# class 확인
tags_ko_all = []
rows = [rows1, rows2]
for idx, r in enumerate(rows):
    for i in range(0, len(r)):
        row = r[i]
        for tag_ko in row[2:]:
            if tag_ko != '':
                tag_ko = preprocess(tag_ko)
                tags_ko_all.append(tag_ko)

print("dataset1 & dataset2")
print("중복 제거되지 않는 tags list :", len(tags_ko_all))
tags_ko_all = sorted(list(set(tags_ko_all)))
print("중복 제거된 tags list ko :", len(tags_ko_all))
```

```python
translator = googletrans.Translator()
tags_en_all = []
for tag_ko in tqdm(tags_ko_all):
    tags_en_all.append(translator.translate(tag_ko, src='ko', dest='en').text)
    time.sleep(0.5)
    
print("중복 제거된 tags list en :", len(tags_en_all))
```

```python
torch.save(tags_ko_all, "./annotation/all_class_list_ko.pt")
torch.save(tags_en_all, "./annotation/all_class_list_en.pt")
```
