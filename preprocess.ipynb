```python
# 모션 제거

import csv
import glob
import os

label_pattern='/home/irteam/junghye-dcloud-dir/Pixt/data/dataset/dataset*/label.csv'
out_pattern='/home/irteam/junghye-dcloud-dir/Pixt/data/dataset/dataset*/motion_excluded.csv'

for i in range(1,3):
    label_paths=glob.glob(label_pattern.replace('*',str(i)))
    out_paths=glob.glob(out_pattern.replace('*',str(i)))

    for label_path,out_path in zip(label_paths,out_paths):
        with open(label_path,'r') as f:
            data=csv.reader(f)
            header=next(data)

            delete_tag='모션'
            new_rows=[]
            for row in data:
                if delete_tag in row:
                    row.remove(delete_tag)
                new_rows.append(row)

            with open(out_path,'w',newline='') as outfile:
                
                outfile.truncate()
                wr=csv.writer(outfile)
                wr.writerows(new_rows)
```
