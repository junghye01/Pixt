# translation file

```python
all_df = pd.read_csv("./annotation/annotation_remove_mgf/annotation_preprocessed.csv")
all_df
```

```python
all_tags_ko_list = []
for index in range(0,all_df.shape[0]):
    tags_ko = eval(all_df.iloc[index]["tags_ko"])
    for tag in tags_ko:
        all_tags_ko_list.append(tag)
all_tags_ko_list = list(set(all_tags_ko_list))
len(all_tags_ko_list)
```

```python
translator = googletrans.Translator()
tmp = {}
for tag in tqdm(all_tags_ko_list):
    tmp[tag] = translator.translate(tag, dest='en').text
    time.sleep(0.5)
```
