```python
import pandas as pd
```

```python
train_df = pd.read_csv("./annotation/annotation_merged_remove_gap.csv")
train_df.shape
```

```python
valid_df = train_df.sample(n=500)
train_df = train_df.drop(valid_df.index, axis=0, inplace=False)
train_df.shape, valid_df.shape
```

```python
train_df = train_df.reset_index(drop=True)
valid_df = valid_df.reset_index(drop=True)
```

```python
train_df.to_csv("./annotation/annotation_merged_remove_gap/train.csv", index=False)
valid_df.to_csv("./annotation/annotation_merged_remove_gap/valid.csv", index=False)
```
