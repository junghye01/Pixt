```python
import csv
import pandas as pd
import torch
```

```python
rows1 = []
# Load dataset1
with open('./dataset1/label.csv', 'r') as f:
    csv_data = csv.reader(f)
    header = next(csv_data)
    rows1 = list(csv_data)

rows2 = []
# Load dataset2
with open('./dataset2/label.csv', 'r') as f:
    csv_data = csv.reader(f)
    header = next(csv_data)
    rows2 = list(csv_data)
```

```python
header
```

```python
# 공백 제거 & 오타 수정
def preprocess(tag_ko):
    tag_ko = tag_ko.replace(" ","")
    tag_ko = tag_ko.replace("ㅐㄴ드마크","랜드마크")
    tag_ko = tag_ko.replace("ㅏㅅ랑","사랑")
    tag_ko = tag_ko.replace("ㅑ경","야경")
    tag_ko = tag_ko.replace("ㅁ나세","만세")
    return tag_ko
```

```python
columns = ["dataset", "image_number", "title", "tags_ko", "motion_exist", "greenscreen_exist"]
df = pd.DataFrame(columns=columns)

tags_length = []
rows = [rows1, rows2]
motion_count = 0
for idx, r in enumerate(rows):
    for i in range(0, len(r)):
        row = r[i]

        tags_ko =[]
        for tag_ko in row[2:]:
            if tag_ko != '':
                tag_ko = preprocess(tag_ko)
                tags_ko.append(tag_ko)
        if "그린스크린" in tags_ko:
            motion_count += 1
        # try: # 모션 제거
        #     tags_ko.remove("모션")
        # except:
        #     pass

        print(i+1, len(tags_ko), tags_ko)
        tags_length.append(len(tags_ko))
        if len(tags_ko) != 0:
            if idx == 0:
                dataset = "dataset1"
            else:
                dataset = "dataset2"
            motion = True if "모션" in row else False
            greenscreen = True if "그린스크린" in row else False
            
            data_to_insert = [
                dataset, # dataset
                row[0], # image_number
                row[1], # title
                tags_ko, # tags_ko
                motion, # motion_exist
                greenscreen, # greenscreen_exist
            ]
            df.loc[len(df)] = data_to_insert

# df.to_csv("./annotation/annotation_merged_remove_gap.csv", index=False)
df
```
